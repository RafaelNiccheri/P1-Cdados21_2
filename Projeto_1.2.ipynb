```python
import io, os, sys, types
from IPython import get_ipython
from nbformat import read
from IPython.core.interactiveshell import InteractiveShell

def find_notebook(fullname, path=None):
    """find a notebook, given its fully qualified name and an optional path

    This turns "foo.bar" into "foo/bar.ipynb"
    and tries turning "Foo_Bar" into "Foo Bar" if Foo_Bar
    does not exist.
    """
    name = fullname.rsplit('.', 1)[-1]
    if not path:
        path = ['']
    for d in path:
        nb_path = os.path.join(d, name + ".ipynb")
        if os.path.isfile(nb_path):
            return nb_path
        # let import Notebook_Name find "Notebook Name.ipynb"
        nb_path = nb_path.replace("_", " ")
        if os.path.isfile(nb_path):
            return nb_path

class NotebookLoader(object):
    """Module Loader for Jupyter Notebooks"""
    def __init__(self, path=None):
        self.shell = InteractiveShell.instance()
        self.path = path

    def load_module(self, fullname):
        """import a notebook as a module"""
        path = find_notebook(fullname, self.path)

        print ("importing Jupyter notebook from %s" % path)

        # load the notebook object
        with io.open(path, 'r', encoding='utf-8') as f:
            nb = read(f, 4)


        # create the module and add it to sys.modules if name in sys.modules:
        #    return sys.modules[name]
        mod = types.ModuleType(fullname)
        mod.__file__ = path
        mod.__loader__ = self
        mod.__dict__['get_ipython'] = get_ipython
        sys.modules[fullname] = mod

        # extra work to ensure that magics that would affect the user_ns
        # actually affect the notebook module's ns
        save_user_ns = self.shell.user_ns
        self.shell.user_ns = mod.__dict__

        try:
          for cell in nb.cells:
            if cell.cell_type == 'code':
                # transform the input to executable Python
                code = self.shell.input_transformer_manager.transform_cell(cell.source)
                # run the code in themodule
                exec(code, mod.__dict__)
        finally:
            self.shell.user_ns = save_user_ns
        return mod


class NotebookFinder(object):
    """Module finder that locates Jupyter Notebooks"""
    def __init__(self):
        self.loaders = {}

    def find_module(self, fullname, path=None):
        nb_path = find_notebook(fullname, path)
        if not nb_path:
            return

        key = path
        if path:
            # lists aren't hashable
            key = os.path.sep.join(path)

        if key not in self.loaders:
            self.loaders[key] = NotebookLoader(path)
        return self.loaders[key]


sys.meta_path.append(NotebookFinder())

from Classes_Funcoes import *
import emoji
import string
import pandas as pd
```

# Naive-Bayes

### Soma das Palavras

```python
Todas_palavras_unicas = qremove_options_unidos[('todos os tweets', '# de ocorrências')].size
```

### Palavras dado que:

#### relevante P(palavra/relevante)

```python
Total_palavras_rel = qremove_options_rel['# de ocorrências'].sum()
Total_palavras_unicas_rel = qremove_options_rel['# de ocorrências'].size  # .reset_index()['index'].value_counts().sum()

qremove_options_rel_com_prob = (qremove_options_unidos['Palavras em tweets relevantes', '# de ocorrências'] + 1) / (
            Total_palavras_rel + Todas_palavras_unicas)


qremove_options_rel_com_prob = pd.DataFrame(qremove_options_rel_com_prob)
qremove_options_rel_com_prob.rename(columns={'# de ocorrências':'prob palavras'}, inplace=True)
qremove_options_rel_com_prob.columns = qremove_options_rel_com_prob.columns.droplevel()
qremove_options_rel_com_prob.sort_values(by='prob palavras', ascending=False)
qremove_options_rel_com_prob = unite_dataframes_col(qremove_options_rel, qremove_options_rel_com_prob,
                                              'Palavras em tweets relevantes', 'qremove_options_rel_prob', replace_nan=0)
qremove_options_rel_com_prob.columns = qremove_options_rel_com_prob.columns.droplevel()
qremove_options_rel_com_prob
```

#### irrelevante P(palavra/irrelevante)

```python
Total_palavras_unicas_irel = qremove_options_irel['# de ocorrências'].size  # reset_index()['index'].value_counts().sum()

Total_palavras_irel = qremove_options_irel['# de ocorrências'].sum()

qremove_options_irel_com_prob = (qremove_options_unidos['Palavras em tweets irrelevantes', '# de ocorrências'] + 1)/(
        Total_palavras_irel + Todas_palavras_unicas)
pd.DataFrame(qremove_options_irel)
qremove_options_irel_com_prob = pd.DataFrame(qremove_options_irel_com_prob)
qremove_options_irel_com_prob.rename(columns={'# de ocorrências':'prob palavras'}, inplace=True)
qremove_options_irel_com_prob.columns = qremove_options_irel_com_prob.columns.droplevel()
qremove_options_irel_com_prob.sort_values(by='prob palavras', ascending=False)
qremove_options_irel_com_prob = unite_dataframes_col(qremove_options_irel, qremove_options_irel_com_prob,
                                              'Palavras em tweets relevantes', 'qremove_options_irel_prob', replace_nan=0)
qremove_options_irel_com_prob.columns = qremove_options_irel_com_prob.columns.droplevel()
qremove_options_irel_com_prob
```

#### DataFrames Com palavras relevantes e irrelevantes juntas

```python
qremove_options_unidos_comprob = unite_dataframes_col(qremove_options_rel_com_prob,
                        qremove_options_irel_com_prob,
                        'Palavras em tweets relevantes', 'Palavras em tweets irrelevantes', replace_nan=0)
qremove_options_unidos_comprob[('todos os tweets', '# de ocorrências')] = qremove_options_unidos_comprob\
[('Palavras em tweets relevantes', '# de ocorrências')] + qremove_options_unidos_comprob\
[('Palavras em tweets irrelevantes', '# de ocorrências')]
qremove_options_unidos_comprob.sort_values(by=('todos os tweets', '# de ocorrências'), ascending=False)
```

#### Todas as palavras P(relevante) e P(irrelevante)

```python
Total_palavras = Total_palavras_irel + Total_palavras_rel
Total_palavras_unicas = qremove_options_unidos_comprob['todos os tweets', '# de ocorrências'].size
prob_rel_qremove_options = Total_palavras_rel/Total_palavras
prob_irel_qremove_options = Total_palavras_irel/Total_palavras
```

```python
print(f'''
{prob_rel_qremove_options * qremove_options_unidos_comprob['Palavras em tweets relevantes', 'prob palavras'].sum()}
{prob_irel_qremove_options * qremove_options_unidos_comprob['Palavras em tweets irrelevantes', 'prob palavras'].sum()}
{prob_rel_qremove_options * qremove_options_unidos_comprob['Palavras em tweets relevantes', 'prob palavras'].sum() + prob_irel_qremove_options * qremove_options_unidos_comprob['Palavras em tweets irrelevantes', 'prob palavras'].sum()}''')
```

```python
lista_relevantes = []
count_relevantes = 0
lista_falso_relevantes = []
count_falso_relevantes = 0
lista_irrelevantes = []
count_irrelevantes = 0
lista_falso_irrelevantes = []
count_falso_irrelevantes = 0

for i, j in zip(dados_teste1['Teste'], dados_teste1['Unnamed: 1']):
    frase = str(i)
    prob_rel = 1
    prob_irel = 1
    separando_emoji = [j for j in frase if j in emoji.UNICODE_EMOJI['en']]
    emoji_pattern = emoji.get_emoji_regexp()
    frase1 = emoji_pattern.sub(r'', frase)
    frase2 = ''.join(ch for chs in frase1 for ch in chs if ch not in list(string.punctuation))
    frase3 = frase2.replace('\n', ' ').split()
    frase = separando_emoji + frase3
    for palavra in frase:
        palavra.lower()
        if palavra in list(qremove_options_unidos_comprob.reset_index()['index']):
            prob_rel *= qremove_options_rel_com_prob['prob palavras'][palavra]
            prob_irel *= qremove_options_irel_com_prob['prob palavras'][palavra]
    P_relevante_frase = prob_rel * prob_rel_qremove_options
    P_irrelevante_frase = prob_irel * prob_irel_qremove_options
    if P_relevante_frase > P_irrelevante_frase:
        classificacao_bot = 1
        if classificacao_bot == j:
            lista_relevantes.append(i)
            count_relevantes += 1
        else:
            lista_falso_relevantes.append(i)
            count_falso_relevantes += 1
    else:
        classificacao_bot = 0
        if classificacao_bot == j:
            lista_irrelevantes.append(i)
            count_irrelevantes += 1
        else:
            lista_falso_irrelevantes.append(i)
            count_falso_irrelevantes += 1
print(f'''lista das frases verdadeiras positivas seguida pelo seu tamanho:
{lista_relevantes}
{count_relevantes}
lista das frases falsas positivas seguida pelo seu tamanho:
{lista_falso_relevantes}
{count_falso_relevantes}
lista das frases verdadeiras negativas seguida pelo seu tamanho:
{lista_irrelevantes}
{count_irrelevantes}
lista das frases falsas negativas seguida pelo seu tamanho:
{lista_falso_irrelevantes}
{count_falso_irrelevantes}''')

print(f'''porcentagem de verdadeiros positivos {count_relevantes}%
porcentagem de falsos positivos {count_falso_relevantes}%
Porcentagem de verdadeiros negativos {count_irrelevantes}%
Porcentagem de falsos negativos {count_falso_irrelevantes}%
Acurácia {(count_relevantes + count_irrelevantes)/2}
''')
```
