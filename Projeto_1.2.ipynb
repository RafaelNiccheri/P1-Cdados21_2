- #### O arquivo Classes_Funcoes foi importado como um arquivo ".ipynb". Para que esse import funcione corretamente foi necess√°ria essa estrutura abaixo, ela basicamente adiciona uma classe que importa notebooks como modules e os trata como ".py"

```python
import time
import io, os, sys, types
from IPython import get_ipython
from nbformat import read
from IPython.core.interactiveshell import InteractiveShell

def find_notebook(fullname, path=None):
    """find a notebook, given its fully qualified name and an optional path

    This turns "foo.bar" into "foo/bar.ipynb"
    and tries turning "Foo_Bar" into "Foo Bar" if Foo_Bar
    does not exist.
    """
    name = fullname.rsplit('.', 1)[-1]
    if not path:
        path = ['']
    for d in path:
        nb_path = os.path.join(d, name + ".ipynb")
        if os.path.isfile(nb_path):
            return nb_path
        # let import Notebook_Name find "Notebook Name.ipynb"
        nb_path = nb_path.replace("_", " ")
        if os.path.isfile(nb_path):
            return nb_path

class NotebookLoader(object):
    """Module Loader for Jupyter Notebooks"""
    def __init__(self, path=None):
        self.shell = InteractiveShell.instance()
        self.path = path

    def load_module(self, fullname):
        """import a notebook as a module"""
        path = find_notebook(fullname, self.path)

        print ("importing Jupyter notebook from %s" % path)

        # load the notebook object
        with io.open(path, 'r', encoding='utf-8') as f:
            nb = read(f, 4)


        # create the module and add it to sys.modules if name in sys.modules:
        #    return sys.modules[name]
        mod = types.ModuleType(fullname)
        mod.__file__ = path
        mod.__loader__ = self
        mod.__dict__['get_ipython'] = get_ipython
        sys.modules[fullname] = mod

        # extra work to ensure that magics that would affect the user_ns
        # actually affect the notebook module's ns
        save_user_ns = self.shell.user_ns
        self.shell.user_ns = mod.__dict__

        try:
          for cell in nb.cells:
            if cell.cell_type == 'code':
                # transform the input to executable Python
                code = self.shell.input_transformer_manager.transform_cell(cell.source)
                # run the code in themodule
                exec(code, mod.__dict__)
        finally:
            self.shell.user_ns = save_user_ns
        return mod

class NotebookFinder(object):
    """Module finder that locates Jupyter Notebooks"""
    def __init__(self):
        self.loaders = {}

    def find_module(self, fullname, path=None):
        nb_path = find_notebook(fullname, path)
        if not nb_path:
            return

        key = path
        if path:
            # lists aren't hashable
            key = os.path.sep.join(path)

        if key not in self.loaders:
            self.loaders[key] = NotebookLoader(path)
        return self.loaders[key]


sys.meta_path.append(NotebookFinder())
from Classes_Funcoes import *
from tqdm.notebook import tqdm
import emoji
import string
import pandas as pd
```

```python
def passo6(carrega):
    inicio = time.time()
    dados6_coleta = {}
    treino_dados6_r = pd.read_excel(f'{Object_of_Study}.xlsx')
    teste_dados6_r = pd.read_excel(f'{Object_of_Study}.xlsx', sheet_name='Teste')
    for passo6_count in tqdm(range(carrega)):
        treino_dados6 = treino_dados6_r
        teste_dados6 = teste_dados6_r
        treino_dados6.rename(columns={'Treinamento':'tweets'}, inplace=True)
        teste_dados6.rename(columns={'Teste':'tweets'}, inplace=True)
        todos_dados = treino_dados6.append(teste_dados6, ignore_index=True)
        treino_dados6 = todos_dados.sample(300)
        teste_dados6 = todos_dados.drop(treino_dados6.index)
        treino_dados6.rename(columns={'tweets':'Treinamento'}, inplace=True)
        teste_dados6.rename(columns={'tweets':'Teste'}, inplace=True)
        dados6_treino_rel = treino_dados6.loc[treino_dados6['Unnamed: 1'] == 1]
        dados6_treino_irel = treino_dados6.loc[treino_dados6['Unnamed: 1'] == 0]
        dados6_treino_rel = dados6_treino_rel.reset_index()
        dados6_treino_irel = dados6_treino_irel.reset_index()
        dados6_treino_rel.drop(columns='index', inplace=True)
        dados6_treino_irel.drop(columns='index', inplace=True)
        rel_passo6 = ListWord(column_name='Treinamento').lists_all(dados6_treino_rel, c_o_c=True, standard_case=True, new_column_name='# de ocorr√™ncias')
        irel_passo6 = ListWord(column_name='Treinamento').lists_all(dados6_treino_irel, c_o_c=True, standard_case=True, new_column_name='# de ocorr√™ncias')
        rel_limp_passo6 = RemoveWords(rel_passo6, '# de ocorr√™ncias', show_removed=False)
        irel_limp_passo6 = RemoveWords(irel_passo6, '# de ocorr√™ncias', show_removed=False)
        qremove_options_rel_passo6 = rel_limp_passo6.remove_options(remove_at={True: '()'}, remove_link={True: '()'},
                                            remove_ponc={True: '()'},remove_laugh={True: '(2, 5)'},
                                            remove_word_sts={True: '(3)'}, remove_num_str={True: '(3, 5)'})
        qremove_options_irel_passo6 = irel_limp_passo6.remove_options(remove_at={True: '()'}, remove_link={True: '()'},
                                            remove_ponc={True: '()'},remove_laugh={True: '(2, 5)'},
                                            remove_word_sts={True: '(3)'}, remove_num_str={True: '(3, 5)'})
        qremove_options_unidos_passo6 = unite_dataframes_col(qremove_options_rel_passo6, qremove_options_irel_passo6,
                                                             'Palavras em tweets relevantes', 'Palavras em tweets irrelevantes', replace_nan=0)
        qremove_options_unidos_passo6[('todos os tweets', '# de ocorr√™ncias')] = qremove_options_unidos_passo6[('Palavras em tweets relevantes', '# de ocorr√™ncias')] + qremove_options_unidos_passo6[('Palavras em tweets irrelevantes', '# de ocorr√™ncias')]



        '''Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes, Naive-Bayes
        '''
        
        Todas_palavras_unicas_passo6 = qremove_options_unidos_passo6[('todos os tweets', '# de ocorr√™ncias')].size
        Total_palavras_rel_passo6 = qremove_options_rel_passo6['# de ocorr√™ncias'].sum()
        Total_palavras_unicas_rel_passo6 = qremove_options_rel_passo6['# de ocorr√™ncias'].size
        qremove_options_rel_com_prob_passo6 = (qremove_options_unidos_passo6['Palavras em tweets relevantes', '# de ocorr√™ncias'] + 1) / (
                    Total_palavras_rel_passo6+ Todas_palavras_unicas_passo6)

        qremove_options_rel_com_prob_passo6 = pd.DataFrame(qremove_options_rel_com_prob_passo6)
        qremove_options_rel_com_prob_passo6.rename(columns={'# de ocorr√™ncias':'prob palavras'}, inplace=True)
        qremove_options_rel_com_prob_passo6.columns = qremove_options_rel_com_prob_passo6.columns.droplevel()
        qremove_options_rel_com_prob_passo6 = unite_dataframes_col(qremove_options_rel_passo6, qremove_options_rel_com_prob_passo6,
                                                      'Palavras em tweets relevantes', 'qremove_options_rel_prob', replace_nan=0)
        qremove_options_rel_com_prob_passo6.columns = qremove_options_rel_com_prob_passo6.columns.droplevel()

        Total_palavras_irel_passo6 = qremove_options_irel_passo6['# de ocorr√™ncias'].sum()
        Total_palavras_unicas_irel_passo6 = qremove_options_irel_passo6['# de ocorr√™ncias'].size
        qremove_options_irel_com_prob_passo6 = (qremove_options_unidos_passo6['Palavras em tweets relevantes', '# de ocorr√™ncias'] + 1) / (
                    Total_palavras_irel_passo6+ Todas_palavras_unicas_passo6)

        qremove_options_irel_com_prob_passo6 = pd.DataFrame(qremove_options_irel_com_prob_passo6)
        qremove_options_irel_com_prob_passo6.rename(columns={'# de ocorr√™ncias':'prob palavras'}, inplace=True)
        qremove_options_irel_com_prob_passo6.columns = qremove_options_irel_com_prob_passo6.columns.droplevel()
        qremove_options_irel_com_prob_passo6 = unite_dataframes_col(qremove_options_irel_passo6, qremove_options_irel_com_prob_passo6,
                                                      'Palavras em tweets relevantes', 'qremove_options_irel_prob', replace_nan=0)
        qremove_options_irel_com_prob_passo6.columns = qremove_options_irel_com_prob_passo6.columns.droplevel()

        qremove_options_unidos_comprob_passo6 = unite_dataframes_col(qremove_options_rel_com_prob_passo6, qremove_options_irel_com_prob_passo6,
                            'Palavras em tweets relevantes', 'Palavras em tweets irrelevantes', replace_nan=0)
        qremove_options_unidos_comprob_passo6[('todos os tweets', '# de ocorr√™ncias')] = qremove_options_unidos_comprob_passo6\
        [('Palavras em tweets relevantes', '# de ocorr√™ncias')] + qremove_options_unidos_comprob_passo6\
        [('Palavras em tweets irrelevantes', '# de ocorr√™ncias')]
        qremove_options_unidos_comprob_passo6.sort_values(by=('todos os tweets', '# de ocorr√™ncias'), ascending=False)

        Total_palavras_passo6 = Total_palavras_irel_passo6 + Total_palavras_rel_passo6
        Total_palavras_unicas_passo6 = qremove_options_unidos_comprob_passo6['todos os tweets', '# de ocorr√™ncias'].size
        prob_rel_qremove_options_passo6 = Total_palavras_rel_passo6/Total_palavras_passo6
        prob_irel_qremove_options_passo6 = Total_palavras_irel_passo6/Total_palavras_passo6
        count_relevantes_passo6 = 0
        count_irrelevantes_passo6 = 0
        for i, j in zip(teste_dados6['Teste'], teste_dados6['Unnamed: 1']):
            frase_passo6 = str(i)
            prob_rel_passo6 = 1
            prob_irel_passo6 = 1
            separando_emoji_passo6 = [j for j in frase_passo6 if j in emoji.UNICODE_EMOJI['en']]
            emoji_pattern_passo6 = emoji.get_emoji_regexp()
            frase1_passo6 = emoji_pattern_passo6.sub(r'', frase_passo6)
            frase2_passo6 = ''.join(ch for chs in frase1_passo6 for ch in chs if ch not in list(string.punctuation))
            frase3_passo6 = frase2_passo6.replace('\n', ' ').split()
            frase_passo6 = separando_emoji_passo6 + frase3_passo6
            for palavra in frase_passo6:
                palavra.lower()
                if palavra in list(qremove_options_unidos_comprob_passo6.reset_index()['index']):
                    prob_rel_passo6 *= qremove_options_rel_com_prob_passo6['prob palavras'][palavra]
                    prob_irel_passo6 *= qremove_options_irel_com_prob_passo6['prob palavras'][palavra]
            P_relevante_frase_passo6 = prob_rel_passo6 * prob_rel_qremove_options_passo6
            P_irrelevante_frase_passo6 = prob_irel_passo6 * prob_irel_qremove_options_passo6
            if P_relevante_frase_passo6 > P_irrelevante_frase_passo6:
                classificacao_bot_passo6 = 1
                if classificacao_bot_passo6 == j:
                    count_relevantes_passo6 += 1
            else:
                classificacao_bot_passo6 = 0
                if classificacao_bot_passo6 == j:
                    count_irrelevantes_passo6 += 1
            dados6_coleta[passo6_count] = (count_relevantes_passo6 + count_irrelevantes_passo6)/2
        passo6_count+=1
    fim = time.time()
    print(f'Todas as iteracoes demorarm {fim-inicio} segundos')
    return dados6_coleta
```
